1. Create a line plot of ZN and INDUS in the housing data.

	a. For ZN, use a solid green line. For INDUS, use a blue dashed line.
    
	b. Change the figure size to a width of 12 and height of 8.
    
	c. Change the style sheet to something you find https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html.


```python
from IPython.display import HTML

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

plt.style.use('fivethirtyeight')
%matplotlib inline
```

```python
# Read in the Boston housing data.
housing_csv = r'C:\Users\HP\Downloads\bostonHousing.csv'
housing = pd.read_csv(housing_csv)
```

```python
housing
```

```python
plt.figure(figsize=(12, 8))
plt.plot(housing['zn'], color='green', linestyle='solid', label='ZN')
plt.plot(housing['indus'], color='blue', linestyle='--', label='INDUS')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Line Plot of ZN and INDUS')
plt.legend()
plt.show()
plt.style.use('dark_background')
```

2. Create a bar chart using col1 and col2 of dummy data.

	a. Give the plot a large title of your choosing.
	b. Move the legend to the lower-left corner.
	c. Do the same thing but with horizontal bars.
	d. Move the legend to the upper-right corner.

```python
df = pd.DataFrame(np.random.randn(10, 4), 
                  columns=['col1', 'col2', 'col3', 'col4'],
                  index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j'])
```

```python
plt.figure(figsize=(8, 6))

plt.bar(df.index, df['col1'], width=0.4, label='Column 1')
plt.bar(df.index, df['col2'], width=0.4, label='Column 2', bottom=df['col1'])

plt.title('Comparison of Column 1 and Column 2')
plt.legend(loc='lower left')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.barh(df.index, df['col1'], height=0.4, label='Column 1')
plt.barh(df.index, df['col2'], height=0.4, label='Column 2', left=df['col1'])
plt.title('Comparison of Column 1 and Column 2 (Horizontal)')
plt.legend(loc='upper right')
plt.show()
```
